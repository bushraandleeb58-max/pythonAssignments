```python
s1 = int(input("Subject 1: "))
s2 = int(input("Subject 2: "))
s3 = int(input("Subject 3: "))
s4 = int(input("Subject 4: "))
s5 = int(input("Subject 5: "))

total = s1 + s2 + s3 + s4 + s5
per = (total / 500) * 100

if per >= 80: grade = "A+"
elif per >= 70: grade = "A"
elif per >= 60: grade = "B"
else: grade = "Fail"

print(f"Total: {total}, Percentage: {per}%, Grade: {grade}")
```
